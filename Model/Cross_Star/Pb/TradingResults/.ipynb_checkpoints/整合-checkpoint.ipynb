```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
Dominant_Info = pd.read_csv('Dominant_CU.csv')
Dominant_Info['start']
Dominant_Info['start'] = [datetime.strptime(it, '%Y/%m/%d') for it in Dominant_Info['start']]
Dominant_Info['end'] = [datetime.strptime(it, '%Y/%m/%d') for it in Dominant_Info['end']]
#Dominant_Info
```

```python
# dat1 = pd.read_csv('CU_all_1_100.csv')
dat2 = pd.read_csv('CU_all_1_200.csv')
dat3 = pd.read_csv('CU_all_1_300.csv')
dat4 = pd.read_csv('CU_all_1_400.csv')
dat5 = pd.read_csv('CU_all_1_500.csv')
dat6 = pd.read_csv('CU_all_1_600.csv')
dat7 = pd.read_csv('CU_all_1_700.csv')
dat8 = pd.read_csv('CU_all_1_800.csv')
dat9 = pd.read_csv('CU_all_1_900.csv')
dat10 = pd.read_csv('CU_all_1_1000.csv')
def sort_dat(dat,bool=True):
    dat['cumpnl'] = np.cumsum(dat['pnl'])
    dat = dat[['平仓时间','cumpnl','合约名称','pnl']]
    dat['合约名称'] = [name[:6] for name in dat['合约名称']]
    dat.drop_duplicates(['平仓时间'],inplace=True)
    dat.columns = ['Date','Cumnl','Name','pnl']
    dat['Date'] = [datetime.strptime(it, '%Y-%m-%d %H:%M:%S') for it in dat['Date']]
    M = len(dat)
    for i in dat.index:
        temp_dat = dat.loc[i]
        temp_name = temp_dat['Name']
        try:
            dominant_name = Dominant_Info.iloc[np.argwhere(Dominant_Info['contract']==temp_name)[0],:]
            if temp_dat['Date']>dominant_name['end'].values[0] or temp_dat['Date']<dominant_name['start'].values[0]:
                dat.drop(i,inplace=True)
        except IndexError:
            dat.drop(i,inplace=True)
            
    if bool:
        return dat[['Date','Cumnl']],[np.mean(dat['pnl']),np.std(dat['pnl'])]
    else:
        return dat,[np.mean(dat['pnl']),np.std(dat['pnl'])]
        
   
```

```python
%matplotlib inline

from vnpy.trader.app.ctaStrategy.ctaBacktesting import BacktestingEngine, OptimizationSetting, MINUTE_DB_NAME, TICK_DB_NAME
from vnpy.trader.app.ctaStrategy.strategy.strategyLimitOrder2Trade import LimitOrder2TradeStrategy
```

```python
# 创建回测引擎对象
engine = BacktestingEngine()
```

```python
# 配置回测引擎参数
engine.setSlippage(0)      # 设置滑点为1跳
engine.setRate(1.05/10000)    # 设置手续费万1
engine.setSize(10)         # 设置合约大小 
engine.setPriceTick(1)     # 设置最小价格变动   
engine.setCapital(5000)   # 设置回测本金
```

```python
# 在引擎中创建策略对象
d = {}                                         # 策略参数配置
engine.initStrategy(LimitOrder2TradeStrategy, d)    # 创建策略对象
```

```python
# 设置回测使用的数据
engine.setBacktestingMode(engine.TICK_MODE)    # 设置引擎的回测模式为K线
```

```python
# 空掉
```

```python
# 设置回测使用的数据
engine.setDatabase(TICK_DB_NAME, 'rb1805')  # 设置使用的历史数据库
engine.setStartDate('20170501')               # 设置回测用的数据起始日期
engine.setEndDate('20170701')               # 设置回测用的数据起始日期
```

```python
# 运行回测
engine.runBacktesting()
```
